### Missing value treatment

```python
# UDF for missing value treatment
def missing_value_imputation(x, stats = 'mean'):
    if (x.dtypes == 'float64') | (x.dtypes == 'int64'):
        x = x.fillna(x.mean()) if stats == 'mean' else x.fillna(x.median())
    else:
        x = x.fillna(x.mode().iloc[0])
    return x
```

### Creating dummies

```python
def create_dummies(dframe, column_name):
    col_dummies = pd.get_dummies(dframe[column_name], prefix = column_name)
    col_dummies.drop(col_dummies.columns[0], axis = 1, inplace = True)
    dframe = pd.concat([dframe, col_dummies], axis = 1)
    dframe.drop(column_name, axis = 1, inplace = True)
    return dframe
```

```python
cred_not_missing_cat=cred_not_missing.select_dtypes(include=['object'])
cat_cols=list(cred_not_missing_cat.columns)
cat_cols
```

## Train and Test split of data

```python
# y variable, also known as target variable is 'cc_cons'. As y variable is numeric, it is a regression problem. 
# All other columns are called x variables.

x_columns=credit_final.columns.difference(['cc_cons'])
x_columns
```

```python
#Splitting the data for y(cc_cons) variable and x variables.
# We are taking test size 0.2 which means train data 80% and test data 20%

from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y=train_test_split(credit_final[x_columns],
                                              credit_final['cc_cons'],
                                              test_size=0.2,
                                              random_state=12)
```

# Modelling used: Random Forest Regression Modelling

```python
import sklearn as sk
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

import statsmodels.formula.api as smf


from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

### Validation using Root Mean Square Percentage Error (RMSPE) 

```python
errors = abs((pred_y - test_y)/test_y)
print('Mean Absolute Percentage Error:', round(np.mean(errors), 5)) 
```

```python
rmspe = np.sqrt(np.mean(np.square(((test_y - pred_y) / test_y)), axis=0))
print('Root Mean Squared Percentage Error (RMSPE)::', rmspe)
```

```python
print('Mean Absolute Percentage Error (MAPE):', metrics.mean_absolute_percentage_error(test_y, pred_y))
```

```python
mape = 100 * (errors / test_y)
mape
```

```python
accuracy = 100 - np.mean(mape)
print('Model Accuracy:', round(accuracy, 2), '%.')
```
